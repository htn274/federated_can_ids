```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
f1_results = pd.DataFrame({
    'Kia': [ 0.9873, 0.9937, 0.9937, 0.9939, 0.9952],
    'Tesla': [ 0.9725, 0.9746, 0.9847, 0.9914, 0.9922],
    'BMW' : [ 0.6918, 0.8117, 0.9263, 0.9387, 0.9806]
}, index =range(int(1e4), int(5e4)+10, int(1e4)))
```

```python
f1_results['Avg'] = f1_results['Kia'] + f1_results['Tesla'] + f1_results['BMW']
f1_results['Avg'] = f1_results['Avg'] / 3
f1_results
```

```python
plt.figure(figsize=(10,5), dpi=300)
ax = sns.lineplot(data=f1_results, markers=True)
plt.setp(ax.get_legend().get_texts(), fontsize='16') # for legend text
```
